```python
# Importar las librerías necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
grammy_df = pd.read_csv('../Data/the_grammy_awards.csv')

# Visualizar las primeras filas del dataset
grammy_df.head()
```

```python
# Información general del dataset
grammy_df.info()

# Revisar si hay valores nulos en el dataset
grammy_df.isnull().sum()

# Estadísticas descriptivas generales
grammy_df.describe(include='all')
```

```python
# Conteo de las nominaciones de cada artista
nominations_by_artist = grammy_df['nominee'].value_counts().head(10)
nominations_by_artist.plot(kind='barh', color='skyblue')
plt.title('Top 10 Nominados de los Grammy', fontsize=16)
plt.xlabel('Cantidad de Nominaciones')
plt.ylabel('Artista')
plt.show()

# Proporción de premios ganados vs. no ganados
plt.figure(figsize=(6, 6))
sns.countplot(data=grammy_df, x='winner', palette='Set2')
plt.title('Distribución de Ganadores y Nominados', fontsize=16)
plt.xlabel('Ganador')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Verificar los valores únicos en la columna 'winner'
print(grammy_df['winner'].value_counts())
```

```python
# Convertir las columnas 'published_at' y 'updated_at' a formato datetime
grammy_df['published_at'] = pd.to_datetime(grammy_df['published_at'], errors='coerce')
grammy_df['updated_at'] = pd.to_datetime(grammy_df['updated_at'], errors='coerce')

# Verificar si hay valores que no pudieron ser convertidos
print(grammy_df[['published_at', 'updated_at']].dtypes)

# Contar los valores NaT (no convertidos)
print(grammy_df[['published_at', 'updated_at']].isna().sum())
```

```python
# Histograma de las fechas de publicación
plt.figure(figsize=(12, 6))
grammy_df['published_at'].dt.year.hist(bins=20, color='lightgreen')
plt.title('Distribución de Fechas de Publicación de los Premios Grammy', fontsize=16)
plt.xlabel('Año')
plt.ylabel('Cantidad de Premios Publicados')
plt.show()

# Histograma de las fechas de actualización
plt.figure(figsize=(12, 6))
grammy_df['updated_at'].dt.year.hist(bins=20, color='lightblue')
plt.title('Distribución de Fechas de Actualización de los Premios Grammy', fontsize=16)
plt.xlabel('Año')
plt.ylabel('Cantidad de Premios Actualizados')
plt.show()
```

```python
# Conteo de premios por año
plt.figure(figsize=(12, 6))
grammy_df['year'].value_counts().sort_index().plot(kind='bar', color='coral')
plt.title('Cantidad de Premios Grammy por Año', fontsize=16)
plt.xlabel('Año')
plt.ylabel('Cantidad de Premios')
plt.show()
```

```python
# Gráfico de caja (boxplot) para el año de los premios
plt.figure(figsize=(10, 6))
sns.boxplot(x=grammy_df['year'], color='lightblue')
plt.title('Detección de Outliers en los Años de Premios', fontsize=16)
plt.show()
```

```python
# Seleccionar las columnas numéricas
numeric_columns = ['year', 'winner']

# Calcular la matriz de correlación
correlation_matrix = grammy_df[numeric_columns].corr()

# Visualizar la matriz de correlación con un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=1, fmt='.2f')
plt.title('Matriz de Correlación de Variables Numéricas (Año y Ganador)', fontsize=16)
plt.show()
```

## Conclusiones

Años y Publicaciones:

La mayor cantidad de premios Grammy se ha publicado en los últimos años, con un aumento significativo en 2019.
La mayoría de los datos de actualización de premios son recientes, concentrándose en un corto periodo de tiempo.
Distribución de Ganadores:

Todos los registros de este dataset son de ganadores (no se incluyen nominaciones que no hayan resultado en un premio). Esto explica que en el gráfico de ganadores y nominados solo haya una barra para "True".
Nominaciones de Artistas:

Los datos muestran a artistas como Bridge Over Troubled Water, Berlioz: Requiem, y Robert Woods entre los más nominados, todos con alrededor de 6-7 nominaciones.
Esto resalta la tendencia de ciertos artistas o producciones a acumular múltiples nominaciones, lo que es común en categorías como "Álbum del Año" o "Productor del Año".
Categorías y Artistas:

La columna de categorías y artistas muestra que los datos están bastante fragmentados, ya que algunos registros no tienen información completa sobre los artistas o trabajadores que estuvieron involucrados, lo que podría requerir limpieza adicional.
Fechas de Publicación y Actualización:

El gráfico de fechas de publicación de premios muestra que hubo una gran cantidad de publicaciones en el año 2019.
El gráfico de actualizaciones está distribuido en periodos no consistentes, lo que podría deberse a la naturaleza del dataset o a la forma en la que los premios Grammy actualizan su base de datos.
Matriz de Correlación:

La matriz de correlación entre año y ganador es mínima, ya que la única variable numérica relevante en este contexto es el año de la premiación, lo cual no presenta mayor variabilidad.
